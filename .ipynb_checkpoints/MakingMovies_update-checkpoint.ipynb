```python
fn = "../{}-{:.4}-FullTestNew/{}-{:.4}-FullTestNew".format(300,8e-5,300,8e-5)
movie_file_name0 = fn + '/' + "f.mp4"
movie_file_name3 = fn + '/' + "n2p.mp4"
    
#if not os.path.isfile(fn+'/movie-arrays.npz'):
    #print("Need to create data arrays.  Will take a few minutes.")
    #create_data_arrays(mH,mixangle,True)
arrs = np.load(fn+'/movie-arrays.npz')
    
hbar = 6.58212e-22
tH = trunc.tH(300,8e-5)*hbar
    
eq_a_array = np.load("../eq_arrs/eq_arrs/Luke_a.npy") #scale factor in standard cosmology
eq_T_array = np.load("../eq_arrs/eq_arrs/Luke_T.npy") #temperature in standard cosmology
eq_eta_array = np.load("../eq_arrs/eq_arrs/Luke_eta.npy") #eta in standard cosmology
    
cs_eta = sp.CubicSpline(np.flip(eq_T_array),np.flip(eq_eta_array))
cs_Tcm = sp.CubicSpline(np.flip(eq_T_array),np.flip(1/eq_a_array))
    
eq_Tcm = cs_Tcm(arrs['T']) 

e_vec = arrs['e'][-1,:]
y_vec = arrs['y'][-1,:]
T = arrs['T'][-1]
t = arrs['t'][-1]
a = arrs['a'][-1]
len_e = len(np.where(e_vec > 0)[0])+1

eps = e_vec[:len_e]
yv = y_vec[:len_e]

#e_thermal = eps**3 / (np.exp(eps / a /  eq_Tcm[-1]) +1) / (2 * np.pi**3)
e_thermal = eps**3 / (np.exp(eps / a /  (T*(4/11)**(1/3))) +1) / (2 * np.pi**3)
nu = eps**3 * yv / (2 * np.pi**3)
#time_text = (r"$T = $"+"{:8.3} MeV".format(T) + '\n' + "$T_{\\rm cm} = $"+"{:8.3} MeV".format(1/a) + '\n' + "t = {:10.3} s".format(t * hbar))

fig0 = plt.figure(figsize=(10,6)) #epsilon vs adjusted occupation fraction
ax0 = plt.axes([0.2,0.12,0.7,0.85])
ax0.loglog(eps, e_thermal, linewidth=2, linestyle='--',color='b')
ax0.loglog(eps, nu, linewidth=2, color='k')
#ax0.text(40,0.01,time_text,fontsize=12)
ax0.set_ylim([1e-10,10])
ax0.set_yticks([1e-10,1e-8,1e-6,1e-4,1e-2,1e0])
ax0.set_xlim([0.5,250])
ax0.set_xticks([1e0,1e1,1e2])
ax0.set_xlabel(r"$\epsilon = E_\nu / T_{cm}$",fontsize=20)
ax0.set_ylabel(r"$\epsilon^3 \, f(\epsilon)$",fontsize=20)
#ax0.set_title("{} MeV, {:.3} s decay lifetime".format(mH,tH))
ax0.tick_params(axis="x", labelsize=14)
ax0.tick_params(axis="y", labelsize=14)
#fig0.savefig("SpectrumShortLived.pdf")
```
